## Introduction
I chose The TMDb Movie Data Set for data analysis. It contains information about 10,000 movies collected from the Movie Database(TMDb), including user ratings and revenue. In my analysis, I would like to check other interesting parts of the dataset.

```python
# I will use this cell to import statements of all the packages I intend to use for the project
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

## Data Wrangling

```python
# Load data. View the first five rows
movies_df=pd.read_csv('tmdb_movies_data.csv')
movies_df.head()
```

## Data Cleaning

```python
# First, I will drop the columns that will not be needed in my analsis.
movies_df.drop(['id', 'imdb_id', 'homepage', 'tagline', 'overview', 'production_companies', 'budget_adj', 'revenue_adj','keywords'], axis=1, inplace=True)
```

```python
# View the first five rows after the modification
movies_df.head()
```

```python
# view the last five rows in the dataset
movies_df.tail()
```

```python
# After showing the last few rows, I realized that some values in the revenue and budget are zeros. In this cell, I decided to drop them.
movies_df.drop(movies_df[movies_df['revenue']==0].index, inplace=True)
movies_df.drop(movies_df[movies_df['budget']==0].index, inplace=True)
movies_df.head()
```

```python
# Check whether the duplicate values exists in the data set
movies_df.duplicated().any()
```

```python
# In this cell, I removed the duplicate values in the dataset
movies_df.drop_duplicates(inplace=True)
```

```python
# i confirmed to determine the duplicates were removed
movies_df.duplicated().any()
```

```python
#checking the missing values in the dataset
movies_df.isnull().sum()
```

```python
# In this cell, I will drop the misssing values
movies_df.dropna(subset=['director', 'cast'], axis=0, inplace=True)
```

```python
#confirm if the missing values were dropped.
movies_df.isnull().sum()
```

```python
# Check the status of the dataset after cleaning by showing the first view rows
movies_df.head(7)
```

Question Two:
    
    What is the bottom 8 movies by popularity?

```python
# check the unpopular value
movies_df.popularity.min()
```

```python
# In this cell, I discovered my data to collect between popularity and movies names.
unpopular_movies=movies_df.pivot_table(values='popularity',index="original_title").sort_values(ascending=True,by='popularity').head(8)
unpopular_movies
```

```python
def label(x,y,t):
    """
    Args:
    x(str):x-axis title
    y(str):y-axis title
    t(str):main title
    
    Return:
    None
    """
    
    plt.xlabel(x)
    plt.ylabel(y)
    plt.title(t)
    plt.show()
```

```python
# visualize the data
unpopular_movies.popularity.plot(kind='bar',fontsize=12,edgecolor='black')
label('movies_title','popularity','Bottom 8 Movies by popularity')
```

 Results
 
-The above analysis shows the bottom 8 movies by popularity

-After the analysis, I was able to find out that Born into the Brothels is the unpopular movie in the world

Question Three:

      What is the top Six revenue movies?

```python
top_revenues = pd.DataFrame(movies_df['revenue'].sort_values(ascending = False))
top_revenues['original_title'] = movies_df['original_title']
data = list(map(str,(top_revenues['original_title'])))
x = list(data[:6])
y = list(top_revenues['revenue'][:6])

# make a plot using pointplot for top 6 revenue movies.
ax = sns.pointplot(x=y,y=x)

#setup the figure size
sns.set(rc={'figure.figsize':(10,5)})
#setup the title and labels of the plot.
ax.set_title("Top 6 revenue Movies",fontsize = 14)
ax.set_xlabel("revenue",fontsize = 12)
sns.set_style("darkgrid")
```

After the analysis above, I was able to realize Avatar is the top most movie by revenue hence the most profitable movie in the world.

Question Four:

     What is the effect of the film budget over time in the first 35 years?

```python
# In this cell, I made pivot table to  connect between year overtime and Budget.
annual_budget=movies_df.pivot_table(values='budget',index='release_year').head(35)
annual_budget
```

```python
# visualize the analysis
annual_budget.plot(kind='line' , fontsize='14')
label('release_year','budget','The degree of budget effect on films over the last 35 years')
```

 **Results**
 
1. The above analysis is a relationship between years and budget.

2. After the analysis, I realized that the buget was lower in the 1960-1975 as compared from 1975 onwards in first 35 years.

3. It is evident the movie industry developed steadly from 1975-1995 hence the increase in budget and productivity.
